## Working with Strings

The data frame contains notes taken by staff members. Your task is to clean it up to make it usable for analysis. 

### Tasks
* Create start date, end date columns and populate the values (If notes don't include month or day use 1) i.e: 2018 => 2018-01-01

* Create phone number column (If notes don't include phone numbers, use None)
* Create a column `duration` showing difference between start & end dates in years as a whole number

```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/px70339sai/DS601_Fall21/main/Week09/data/hw_data_less_spicy.csv', index_col=0)
```

```python
df.head()
```

```python
df['StartDate'] = None
df['EndDate'] = None
df['Duration'] = None
df
```

```python
df.columns
```

```python
phoneNumber = re.compile(r'\(?([0-9]{3})\)?([ .-]?)([0-9]{3})([ .-]?)([0-9]{4})(x([0-9]{1,}))?')
```

```python
df['phone_number'] = [phoneNumber.search(x).group() if phoneNumber.search(str(x)) !=None else None for x in df['notes']]
df.head()
```

```python
dates=[]
for i in range(0,len(df['notes'])):
    dates_list = [x.group() for x in re.finditer( r'([0-9]{2}(-|\/)[0-9]{2}(-|\/)[0-9]{4})|([0-9]{4}(-|\/)[0-9]{2}(-|\/)[0-9]{2})|((January|February|March|April|May|June|July|August|September|October|November|December) [0-9]{4})|([0-9]{2}(\/)[0-9]{2})|(\b[0-9]{4}\b)(-[0-9]{2})?', df['notes'][i])]
    dates.append(dates_list)
df['dates']=dates
df.head()
        
```

```python

df['temp_start_date']= [date[0] for date in df['dates']]
df['temp_end_date'] = [date[1] for date in df['dates']]

df.head()
```

```python
df['StartDate'] = pd.to_datetime(df['temp_start_date'], errors='ignore')
df['EndDate'] = pd.to_datetime(df['temp_end_date'], errors='ignore')

df.drop('temp_start_date', axis=1, inplace=True)
df.drop('temp_end_date', axis=1, inplace=True)
df.drop('dates', axis=1, inplace=True)

df.head()
```

```python
df['Duration'] = df['EndDate'] - df['StartDate']

df['Duration'] = df['Duration'] / np.timedelta64(1, 'Y')
df.head()
```
