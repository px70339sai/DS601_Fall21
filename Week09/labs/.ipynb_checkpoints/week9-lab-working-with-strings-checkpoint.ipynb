__Your Task__

- In  `Q12_Part_4_TEXT` they ask the tools the participants are using day to day. However as you might observe the answers are not standardized. 

- Find all the answers with 'jupyter' mentioned in it.

- Find all the participant who are Python users.

- Find all R users. 

```python
import re
```

```python
#pattern.search((str(s)) == None else False for s in df['Q12_Part_4_TEXT']
```

```python
 df['Q12_Part_4_TEXT']
```

```python
pattern = re.compile(r'Jupyter', re.I)
```

```python
df['mentioned_jupyter']= [True if pattern.search(str(s))!= None else False for s in df['Q12_Part_4_TEXT']]
```

```python
df.head()
```

```python
df[df['mentioned_jupyter']==True]['Q12_Part_4_TEXT']
```

```python
pattern = re.compile(r'Python', re.I)
```

```python
df['mentioned_Python']= [True if pattern.search(str(s))!= None else False for s in df['Q12_Part_4_TEXT']]
```

```python
df.head()
```

```python
df[df['mentioned_Python']==True]['Q12_Part_4_TEXT']
```
